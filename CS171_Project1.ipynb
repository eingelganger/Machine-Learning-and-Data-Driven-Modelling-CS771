```python
def transform(X):
  row = len(X)
  col = len(X[0])
  X_transform = np.zeros((row,col*10))
  for j in range(row):
    s = X[j]
    s_transform = np.zeros((col*10,1))
    i=0
    while (s[i] == '0'):
      s_transform[i*10:(i+1)*10,:] = np.zeros((10,1))
      i+=1
    while (i<col):
      s_transform[i*10:(i+1)*10,:] = np.eye(N = 10,M = 1,k=-int(s[i]))
      i+=1
    X_transform[j,:] = s_transform.T
  return X_transform
```

```python
train_seq_X_transform = []
for s in train_seq_X:
  s_transform = []
  for i in range(0,50,1):
    s_transform = s_transform + [s[i]]
  for i in range(0,49,2):
    s_transform = s_transform + [s[i:i+2]]
  for i in range(0,48,3):
    s_transform = s_transform + [s[i:i+3]]
  train_seq_X_transform.append(s_transform)

train_seq_X_transform = np.array(train_seq_X_transform)
```

```python
categorical_features_indices = list(range(train_seq_X_transform.shape[1]))

cb_model = CatBoostClassifier(iterations=400, learning_rate=0.1, depth=12,verbose=False)

cb_model.fit(train_seq_X_transform, train_seq_Y,cat_features=categorical_features_indices)
```

```python
importances = cb_model.feature_importances_

# Select features based on importance threshold
important_features = [i for i, imp in enumerate(importances) if imp > 0.001]
train_seq_X_transform_new = train_seq_X_transform[important_features,:]
train_seq_Y_new = np.array(train_seq_Y)
train_seq_Y_new = train_seq_Y_new[important_features]
```
